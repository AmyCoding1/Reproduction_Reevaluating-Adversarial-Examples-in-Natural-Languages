```python
def fix(x):
    return 'Real' if x else 'Computer-altered'

df['Orig'] = df['Input.original'].apply(fix)
```

```python
df['Correct'] = df['Orig'] == df['Answer.sentiment.label']
len(df[df['Correct']])
```

```python
df_corr = df[df['Correct']]
```

```python
df_sus_corr = df_corr[df_corr['Orig'] == 'Computer-altered']
len(df_sus_corr)
```
