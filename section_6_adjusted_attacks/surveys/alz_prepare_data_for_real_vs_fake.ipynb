```python
import pandas as pd

df = pd.read_csv('alz_mr_examples.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df = df.rename(columns={'passage_1': 'text1', 'passage_2': 'text2'})
```

```python
len(df)
```

```python
# We want labels on 88 samples to match the number of labels we got on results running TextFooler against BERT MR.
# Calling sample() shuffles the dataframe before taking the first 88 rows.
sample_df = df.sample(frac=1)[:88]
sample_df.head()
```

```python
# Now make a new df with 'real' and 'fake' data.
orig_df = sample_df['text1'].to_frame()
orig_df = orig_df.rename(columns={'text1': 'text'})
orig_df['label'] = 'real'

fake_df = sample_df['text2'].to_frame()
fake_df = fake_df.rename(columns={'text2': 'text'})
fake_df['label'] = 'fake'

mturk_df = pd.concat((orig_df, fake_df)).sample(frac=1).reset_index().drop('index', axis=1)
mturk_df.head()
```

```python
print(mturk_df['label'].value_counts())
```

```python
mturk_df.to_csv('alz_s5_176_samples_for_mturk.csv', index=False)
```
